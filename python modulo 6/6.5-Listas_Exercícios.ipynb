````python
### crie a função ordena_por_comprimento e retorne a lista resultado


###
def ordena_por_comprimento(lista):
    """
    Ordena a lista de strings com base no comprimento das strings.
    :param lista: Lista de strings a ser ordenada
    :return: Lista ordenada por comprimento das strings
    """
    return sorted(lista, key=lambda s: len(s))

# Lista de nomes
nomes = ["Joao", "Maria", "Jose", "Gabriela", "Sol", "Luna", "Bento", "Enzo", "Fernanda"]

# Aplicando a função
nomes_ordenados = ordena_por_comprimento(nomes)

# Imprimindo o resultado
print(nomes_ordenados)

````

### Q5.

Escreva um programa para encontrar a diferença entre duas listas, incluindo elementos duplicados.

Ex:
```
Listas originais: 
[1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
[1, 1, 2, 4, 5, 6]
Diferença entre as listas:
[3, 3, 4, 7]
```

````python
# Implemente sua solução aqui
from collections import Counter

def diferença_listas(lista1, lista2):
    """
    Encontra a diferença entre duas listas, incluindo elementos duplicados.
    :param lista1: Primeira lista
    :param lista2: Segunda lista
    :return: Lista contendo a diferença, com elementos duplicados
    """
    # Conta a frequência de cada elemento em ambas as listas
    contagem1 = Counter(lista1)
    contagem2 = Counter(lista2)
    
    # Cria uma lista para armazenar a diferença
    diferença = []
    
    # Itera sobre os elementos e suas contagens na primeira lista
    for elem in contagem1:
        if elem in contagem2:
            # Se o elemento está em ambas as listas, calcula a diferença de contagens
            diferença.extend([elem] * (contagem1[elem] - contagem2[elem]))
        else:
            # Se o elemento não está na segunda lista, adiciona todas as ocorrências
            diferença.extend([elem] * contagem1[elem])
    
    return diferença

# Listas de exemplo
lista1 = [1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
lista2 = [1, 1, 2, 4, 5, 6]

# Encontrar a diferença entre as listas
resultado = diferença_listas(lista1, lista2)
print("Diferença entre as listas:")
print(resultado)  # Saída esperada: [3, 3, 4, 7]
````

Você está desenvolvendo um programa para determinar o vencedor de um jogo da velha. Escreva uma função ```avalia_tabuleiro()``` que recebe uma lista representando o estado atual do tabuleiro e determina se há um vencedor. 

A lista terá 3 sublistas, cada uma representando uma linha do tabuleiro. Cada elemento das sublistas pode ser 'X', 'O' ou vazio (' ').

O programa deve imprimir o vencedor ('X', 'O'), se houver um, ou "Empate" se não houver vencedor. Um vencedor é determinado se houver três 'X' ou três 'O' em linha (horizontal, vertical ou diagonal).

Exemplo:
```python
tabuleiro = [
    ['X', 'O', 'X'],
    [' ', 'X', 'O'],
    ['O', ' ', 'O']
]
```
O resultado seria "Empate", pois não há um vencedor no exemplo fornecido. 

Exemplo 2:
```python
tabuleiro = [
    ['O', 'X', 'O'],
    ['X', 'O', 'X'],
    ['X', ' ', 'O']
]
```
O resultado seria "O", pois 'O' venceu na diagonal central.

````python
### Escreva a função avalia_tabuleiro()




### teste a função nos seguintes tabuleiros

tabuleiro1 = [
    ['X', 'O', 'X'],
    [' ', 'X', 'O'],
    ['O', ' ', 'X']
]

tabuleiro2 = [
    ['X', 'O', 'O'],
    ['X', 'X', 'O'],
    ['X', 'O', 'X']
]

tabuleiro3 = [
    [' ', ' ', ' '],
    ['X', ' ', 'O'],
    ['O', 'X', 'X']
]

tabuleiro4 = [
    ['O', 'O', 'O'],
    ['X', ' ', 'X'],
    ['O', 'X', 'X']
]

tabuleiro5 = [
    ['X', 'X', 'O'],
    ['X', ' ', 'O'],
    ['O', 'O', 'X']
]


def avalia_tabuleiro(tabuleiro):
    """
    Avalia o estado do tabuleiro de jogo da velha para determinar o vencedor.
    :param tabuleiro: Lista de listas representando o tabuleiro (3x3)
    :return: O vencedor ('X' ou 'O') ou 'Empate' se não houver vencedor
    """
    # Função auxiliar para verificar se todos os valores em uma lista são iguais e não estão vazios
    def todos_iguais(l):
        return l[0] != ' ' and l[0] == l[1] == l[2]

    # Verificar linhas horizontais
    for linha in tabuleiro:
        if todos_iguais(linha):
            return linha[0]

    # Verificar linhas verticais
    for col in range(3):
        if todos_iguais([tabuleiro[0][col], tabuleiro[1][col], tabuleiro[2][col]]):
            return tabuleiro[0][col]

    # Verificar diagonais
    if todos_iguais([tabuleiro[0][0], tabuleiro[1][1], tabuleiro[2][2]]):
        return tabuleiro[0][0]
    if todos_iguais([tabuleiro[0][2], tabuleiro[1][1], tabuleiro[2][0]]):
        return tabuleiro[0][2]

    # Se não há vencedor, retornar 'Empate'
    return 'Empate'

# Exemplo de uso
tabuleiro1 = [
    ['X', 'O', 'X'],
    ['O', 'X', 'O'],
    ['O', 'X', 'X']
]

tabuleiro2 = [
    ['X', 'O', ' '],
    ['O', ' ', 'X'],
    [' ', 'O', 'X']
]

tabuleiro3 = [
    ['X', ' ', ' '],
    [' ', 'X', ' '],
    [' ', ' ', 'X']
]

print(avalia_tabuleiro(tabuleiro1))  # Saída esperada: X
print(avalia_tabuleiro(tabuleiro2))  # Saída esperada: Empate
print(avalia_tabuleiro(tabuleiro3))  # Saída esperada: X
````
