### Q2.

Dado um número inteiro positivo informado pelo usuário, crie um programa em Python que utilize o comando for para calcular e exibir a soma dos números de 1 até o número informado. Ou seja, $1 + 2 + 3 + ... + n$

Exemplo de interação:
```
Digite um número: 5
A soma dos números de 1 a 5 é 15
```


````python
## Escreva e execute seu código aqui
# Solicita ao usuário que informe um número inteiro positivo
numero = int(input("Digite um número inteiro positivo: "))

# Inicializa a variável que armazenará a soma
soma = 0

# Itera sobre todos os números de 1 até o número informado
for i in range(1, numero + 1):
    soma += i

# Exibe o resultado
print(f"A soma dos números de 1 até {numero} é {soma}.")
````

### Q3.

Escreva um programa que lê 10 valores inteiros positivos do usuário e ao final imprime a média dos valores digitados com duas casas decimais.

```
Digite 10 números positivos:
3
2
12
14
20
42
19
0
8
1
A média dos valores digitados é 12.10
```

````python
## Escreva e execute seu código aqui
# Inicializa a soma dos valores e o contador
soma = 0

# Lê 10 valores inteiros positivos do usuário
for i in range(10):
    while True:
        try:
            valor = int(input(f"Digite o valor {i+1} (inteiro positivo): "))
            if valor <= 0:
                raise ValueError("O valor deve ser um número inteiro positivo.")
            break
        except ValueError as e:
            print(e)
    
    soma += valor

# Calcula a média
media = soma / 10

# Imprime a média com duas casas decimais
print(f"A média dos valores digitados é: {media:.2f}")
````

### Q4.

Você vai criar um sistema que registra os resultados dos jogos do Atlético MG ao longo de um campeonato. Seu sistema vai receber os resultados de todos os jogos do Galo, e deve calcular a pontuação do time sabendo que vitórias valem 3 pontos, empates 1 ponto e derrotas 0 pontos.

Entrada: <br>
A primeira linha de entrada é um inteiro N com a quantidade jogos do galo. Para cada jogo você deve ler 2 inteiros, o primeiro com a quantidade de gols do galo e o segundo com a quantidade de gols do time oponente. 

Saída: <br>
Apresente a soma de vitórias, empates e derrotas do galo, junto com o cálculo da pontuação total.

Ex:
```
Entrada:
4
2
0
1
1
0
3
1
0

Saída:
Vitórias: 2
Empates: 1
Derrotas: 1
Pontuação: 7

```




````python
## Escreva e execute seu código aqui
# Lê o número de jogos
N = int(input("Digite o número de jogos: "))

# Inicializa contadores
vitorias = 0
empates = 0
derrotas = 0
pontuacao_total = 0

# Processa cada jogo
for _ in range(N):
    # Lê os gols do Atlético MG e do time oponente
    gols_atletico, gols_oponente = map(int, input().split())

    # Determina o resultado do jogo
    if gols_atletico > gols_oponente:
        vitorias += 1
        pontuacao_total += 3
    elif gols_atletico == gols_oponente:
        empates += 1
        pontuacao_total += 1
    else:
        derrotas += 1

# Exibe o resultado
print(f"Vitórias: {vitorias}")
print(f"Empates: {empates}")
print(f"Derrotas: {derrotas}")
print(f"Pontuação total: {pontuacao_total}")
````

### Q5.

Faça um programa que lê dois inteiros N e M, e imprime na tela um campo de batalha naval. O tabuleiro deve possuir N linhas e M colunas. A primeira linha é composta por um espaço em branco e o cabeçalho das colunas, ou seja, valores de 1 a M. As N linhas seguintes iniciam com o cabeçalho da linha, ou seja, seu número, seguido de M caracteres "/" (barra) indicando uma possível posição jogável. 

```
Entrada:
5
4

Saída:
  1 2 3 4 
1 / / / / 
2 / / / / 
3 / / / / 
4 / / / / 
5 / / / / 
```

Para esse exercício, precisamos lembrar que o comando print implicitamente adiciona uma quebra de linha ao final da impressão. Podemos interferir no final da impressão adicionando mais uma entrada ao print. No exemplo, finalizamos cada linha com um espaço em branco:
```python
print("Texto qualquer", end = " ")
```

````python
## Escreva e execute seu código aqui
# Lê os inteiros N e M
N = int(input("Digite o número de linhas (N): "))
M = int(input("Digite o número de colunas (M): "))

# Imprime o cabeçalho das colunas
# Primeiro imprime um espaço em branco seguido pelos números das colunas
print(" " + " ".join(str(i) for i in range(1, M + 1)))

# Imprime as linhas do campo de batalha
for i in range(1, N + 1):
    # Imprime o número da linha seguido por M barras
    print(f"{i} " + " ".join("/" * M))
````

### Q6.

Escreva um programa que lê uma quantidade indefinida de valores e informa o maior e o menor valor digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
12
-1
29
11
-5
0

Saída:
Maior: 29
Menor: -5
```

````python
## Escreva e execute seu código aqui
# Inicializa as variáveis para armazenar o maior e o menor valor
maior = None
menor = None

print("Digite os valores. A leitura será encerrada quando você digitar 0.")

while True:
    # Lê o valor do usuário
    valor = int(input("Digite um valor: "))
    
    # Encerra o loop se o valor for 0
    if valor == 0:
        break
    
    # Atualiza o maior e o menor valor
    if maior is None or valor > maior:
        maior = valor
    if menor is None or valor < menor:
        menor = valor

# Verifica se algum valor foi digitado (exceto o 0 de término)
if maior is not None and menor is not None:
    print(f"O maior valor digitado foi: {maior}")
    print(f"O menor valor digitado foi: {menor}")
else:
    print("Nenhum valor válido foi digitado.")
````

### Q7.

Escreva um programa que lê uma quantidade indefinida de valores e informa o **produto** dos valores positivos digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
2
-1
1
3
-5
0

Saída:
Produto: 6
```

````python
## Escreva e execute seu código aqui
# Inicializa o produto dos valores positivos
produto = 1
encontrou_positivo = False

print("Digite os valores. A leitura será encerrada quando você digitar 0.")

while True:
    # Lê o valor do usuário
    valor = int(input("Digite um valor: "))
    
    # Encerra o loop se o valor for 0
    if valor == 0:
        break
    
    # Atualiza o produto se o valor for positivo
    if valor > 0:
        produto *= valor
        encontrado_positivo = True

# Verifica se algum valor positivo foi digitado
if encontrado_positivo:
    print(f"O produto dos valores positivos digitados é: {produto}")
else:
    print("Nenhum valor positivo foi digitado.")
````

### Q8.

Vamos fazer uma calculadora que aceita expressões aritméticas de qualquer tamanho até que o usuário digite a palavra "Fim". Sua calculadora executa apenas as operações "+", "-".

```
Entrada:
5
+
8
-
3
+
2
Fim

Saída:
12
```

````python
## Escreva e execute seu código aqui
# Função para calcular uma expressão aritmética
def calcular_expressao(expressao):
    try:
        # Avalia a expressão aritmética
        resultado = eval(expressao)
        # Retorna o resultado
        return resultado
    except Exception as e:
        # Retorna um erro se a expressão for inválida
        return f"Erro: {e}"

print("Calculadora simples. Digite suas expressões aritméticas ou 'Fim' para encerrar.")

while True:
    # Lê a expressão do usuário
    expressao = input("Digite uma expressão: ")
    
    # Encerra o loop se o usuário digitar 'Fim'
    if expressao.strip().lower() == "fim":
        break
    
    # Calcula e exibe o resultado
    resultado = calcular_expressao(expressao)
    print(f"Resultado: {resultado}")

print("Calculadora encerrada.")
````
