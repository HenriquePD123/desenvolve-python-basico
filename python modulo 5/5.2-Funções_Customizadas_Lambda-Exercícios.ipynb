### Q1.

- Sabendo que o código a seguir calcula o fatorial de n, escreva uma função chamada ```fatorial()``` que recebe um inteiro ```n``` como parâmetro e retorna o resultado do fatorial de ```n```. 
- No programa principal, peça ao usuário o valor de ```n```, chame a sua função e imprima o retorno.

```python
fat = 1
for i in range(1, n+1):
    fat *= i 
```

````python
## Crie aqui a função fatorial()
## lembre-se de executar a célula para que 
## a função passe a existir e possa ser chamada
````

````python
## Escreva aqui o programa principal e execute a célula
def fatorial(n):
    """
    Calcula o fatorial de n.
    
    Args:
    n (int): O número para calcular o fatorial.
    
    Returns:
    int: O fatorial de n.
    """
    # Inicializa o resultado do fatorial
    fat = 1
    
    # Calcula o fatorial usando um loop
    for i in range(1, n + 1):
        fat *= i
    
    return fat

def main():
    # Solicita ao usuário o valor de n
    n = int(input("Digite um número inteiro não-negativo para calcular o fatorial: "))
    
    # Verifica se o número é não-negativo
    if n < 0:
        print("O número deve ser não-negativo.")
        return
    
    # Calcula o fatorial usando a função fatorial
    resultado = fatorial(n)
    
    # Exibe o resultado
    print(f"O fatorial de {n} é {resultado}.")

if __name__ == "__main__":
    main()

````

### Q2.

- Escreva uma função em Python chamada ```soma_quadrados``` que recebe dois números como parâmetros e retorna a soma dos seus quadrados. 
- No programa principal solicite ao usuário que insira dois números e utilize a função para exibir a soma dos quadrados.

````python
## Crie aqui a função soma_quadrados()
````

````python
## Escreva aqui o programa principal e execute a célula
def soma_quadrados(a, b):
    """
    Calcula a soma dos quadrados de dois números.
    
    Args:
    a (int or float): O primeiro número.
    b (int or float): O segundo número.
    
    Returns:
    float: A soma dos quadrados de a e b.
    """
    return a**2 + b**2

def main():
    # Solicita ao usuário que insira dois números
    try:
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))
    except ValueError:
        print("Por favor, insira números válidos.")
        return
    
    # Calcula a soma dos quadrados usando a função soma_quadrados
    resultado = soma_quadrados(num1, num2)
    
    # Exibe o resultado
    print(f"A soma dos quadrados de {num1} e {num2} é {resultado:.2f}.")

if __name__ == "__main__":
    main()
````

### Q3.

- Crie uma função em Python chamada ```soma_digitos``` que recebe um número inteiro como parâmetro e retorna a soma dos seus dígitos. Por exemplo, para o número 123, a função deve retornar 6, $(1 + 2 + 3)$.
    - O desafio aqui é separar os dígitos de um número inteiro usando operações aritméticas <br> <br>

- No programa principal solicite ao usuário que insira um número e utilize a função ```soma_digitos``` para calcular e exibir a soma dos seus dígitos.

````python
## Crie aqui a função soma_digitos()
````

````python
## Escreva aqui o programa principal e execute a célula
def soma_digitos(numero):
    """
    Calcula a soma dos dígitos de um número inteiro.
    
    Args:
    numero (int): O número para calcular a soma dos dígitos.
    
    Returns:
    int: A soma dos dígitos do número.
    """
    soma = 0
    # Usa o valor absoluto para lidar com números negativos
    numero = abs(numero)
    
    while numero > 0:
        # Obtém o último dígito do número
        digito = numero % 10
        # Adiciona o dígito à soma
        soma += digito
        # Remove o último dígito do número
        numero //= 10
    
    return soma

def main():
    # Solicita ao usuário que insira um número inteiro
    try:
        numero = int(input("Digite um número inteiro: "))
    except ValueError:
        print("Por favor, insira um número inteiro válido.")
        return
    
    # Calcula a soma dos dígitos usando a função soma_digitos
    resultado = soma_digitos(numero)
    
    # Exibe o resultado
    print(f"A soma dos dígitos de {numero} é {resultado}.")

if __name__ == "__main__":
    main()
````

### Q4.

- Crie a função ```inverteValor()``` que recebe um inteiro de qualquer tamanho e retorna esse valor invertido usando apenas operações aritméticas
- Crie a função ```verificaInverso()``` que recebe o valor original e o valor invertido e retorna verdadeiro se ambos forem igualmente par ou igualmente ímpar. Retorne falso caso contrário.
- No programa principal, peça um valor do usuário e imprima o retorno de ambas as funções.


````python
## Crie aqui as funções inverteValor() e verificaInverso()

````

````python
## Escreva aqui o programa principal e execute a célula
def inverteValor(numero):
    """
    Inverte os dígitos de um número inteiro.
    
    Args:
    numero (int): O número a ser invertido.
    
    Returns:
    int: O número invertido.
    """
    invertido = 0
    numero_abs = abs(numero)  # Trabalha com o valor absoluto do número
    
    while numero_abs > 0:
        # Obtém o último dígito do número
        digito = numero_abs % 10
        # Adiciona o dígito invertido ao resultado
        invertido = invertido * 10 + digito
        # Remove o último dígito do número
        numero_abs //= 10
    
    # Se o número original for negativo, o resultado invertido deve ser negativo
    if numero < 0:
        invertido = -invertido
    
    return invertido

def verificaInverso(original, invertido):
    """
    Verifica se o número original e o número invertido são ambos pares ou ambos ímpares.
    
    Args:
    original (int): O número original.
    invertido (int): O número invertido.
    
    Returns:
    bool: True se ambos forem igualmente pares ou igualmente ímpares, False caso contrário.
    """
    # Verifica se ambos são pares ou ambos são ímpares
    return (original % 2 == invertido % 2)

def main():
    # Solicita ao usuário que insira um número inteiro
    try:
        numero = int(input("Digite um número inteiro: "))
    except ValueError:
        print("Por favor, insira um número inteiro válido.")
        return
    
    # Calcula o número invertido usando a função inverteValor
    numero_invertido = inverteValor(numero)
    
    # Verifica se o original e o invertido são ambos pares ou ambos ímpares
    resultado_verificacao = verificaInverso(numero, numero_invertido)
    
    # Exibe o resultado
    print(f"O número invertido é: {numero_invertido}")
    print(f"Os números são igualmente pares ou igualmente ímpares? {'Sim' if resultado_verificacao else 'Não'}.")

if __name__ == "__main__":
    main()
````

### Q5.

Você está desenvolvendo um programa para auxiliar em cálculos de geometria básica. Crie as seguintes funções:
- A função ```calcula_perimetro_triangulo()``` que recebe três inteiros correspondentes aos lados de um triângulo e retorna o perímetro do triângulo, ou seja, a soma dos seus lados.
- A função ```calcula_perimetro_circulo()``` que recebe um inteiro referente ao raio do círculo e retorna o perímetro do círculo, dado por $2 \pi r$. Use a constante $\pi$ da biblioteca ```math```.
- A função ```calcula_perimetro_retangulo()``` que possui um parâmetro obrigatório ```lado1``` e um opcional ```lado2```, ambos inteiros. Se o valor opcional não for fornecido, significa que se trata de um quadrado. Sua função deve calcular e retornar o perímetro do retângulo, ou seja, a soma de seus lados. 
    - Para o quadrado, é dado por $4 \times lado1$
    - Para o retângulo é dado por $2 \times lado1 + 2 \times lado2$
    
- No programa principal apresente um menu com as opções disponíveis do seu sistema e uma quarta opção ```Sair```. Solicite ao usuário a opção desejada, solicite as entradas correspondentes à opção escolhida, invoque a respective função e apresente o seu retorno. Seu programa deve retornar ao menu até que o usuário escolha a opção ```Sair```

```
3 - Calcular perímetro retângulo

```

````python
## Crie aqui as funções 

````

````python
## Escreva aqui o programa principal 
import math

def calcula_perimetro_triangulo(lado1, lado2, lado3):
    """
    Calcula o perímetro de um triângulo dado os três lados.
    
    Args:
    lado1 (int): O comprimento do primeiro lado.
    lado2 (int): O comprimento do segundo lado.
    lado3 (int): O comprimento do terceiro lado.
    
    Returns:
    int: O perímetro do triângulo.
    """
    return lado1 + lado2 + lado3

def calcula_perimetro_circulo(raio):
    """
    Calcula o perímetro (circunferência) de um círculo dado o raio.
    
    Args:
    raio (int): O raio do círculo.
    
    Returns:
    float: O perímetro do círculo.
    """
    return 2 * math.pi * raio

def calcula_perimetro_retangulo(lado1, lado2=None):
    """
    Calcula o perímetro de um retângulo. Se lado2 não for fornecido, calcula o perímetro de um quadrado.
    
    Args:
    lado1 (int): O comprimento de um dos lados do retângulo ou quadrado.
    lado2 (int, opcional): O comprimento do outro lado do retângulo. Se não fornecido, considera-se um quadrado.
    
    Returns:
    int: O perímetro do retângulo ou quadrado.
    """
    if lado2 is None:
        # Se lado2 não for fornecido, é um quadrado
        return 4 * lado1
    else:
        # Se lado2 for fornecido, é um retângulo
        return 2 * (lado1 + lado2)

def main():
    while True:
        # Exibe o menu
        print("\nMenu:")
        print("1. Calcular perímetro do triângulo")
        print("2. Calcular perímetro do círculo")
        print("3. Calcular perímetro do retângulo/quadrado")
        print("4. Sair")
        
        # Solicita a opção do usuário
        opcao = input("Escolha uma opção (1/2/3/4): ")
        
        if opcao == '1':
            # Calcula o perímetro do triângulo
            lado1 = int(input("Digite o comprimento do primeiro lado: "))
            lado2 = int(input("Digite o comprimento do segundo lado: "))
            lado3 = int(input("Digite o comprimento do terceiro lado: "))
            resultado = calcula_perimetro_triangulo(lado1, lado2, lado3)
            print(f"O perímetro do triângulo é: {resultado}")
        
        elif opcao == '2':
            # Calcula o perímetro do círculo
            raio = int(input("Digite o raio do círculo: "))
            resultado = calcula_perimetro_circulo(raio)
            print(f"O perímetro do círculo é: {resultado:.2f}")
        
        elif opcao == '3':
            # Calcula o perímetro do retângulo/quadrado
            lado1 = int(input("Digite o comprimento do primeiro lado: "))
            lado2 = input("Digite o comprimento do segundo lado (deixe em branco se for um quadrado): ")
            if lado2 == "":
                resultado = calcula_perimetro_retangulo(lado1)
            else:
                lado2 = int(lado2)
                resultado = calcula_perimetro_retangulo(lado1, lado2)
            print(f"O perímetro do retângulo/quadrado é: {resultado}")
        
        elif opcao == '4':
            # Sair do programa
            print("Saindo...")
            break
        
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
````

### Q6.

Crie uma função lambda para verificar se um número é par ou ímpar. Em seguida, solicite ao usuário um número indefinidos de valores (até que o usuário digite 0). Para cada valor de entrada, informe se é par ou ímpar.

Exemplo de interação:
```
Digite os valores que deseja verificar a paridade (digite 0 para finalizar a entrada de dados):
3
ímpar
8
par
12
par
5
ímpar
7ímpar
```

````python
# Define a função lambda para verificar se um número é par ou ímpar
verifica_par_ou_impar = lambda x: "par" if x % 2 == 0 else "ímpar"

def main():
    while True:
        try:
            # Solicita ao usuário que insira um número
            numero = int(input("Digite um número (0 para sair): "))
            
            if numero == 0:
                # Encerra o loop se o número for 0
                print("Saindo...")
                break
            
            # Verifica se o número é par ou ímpar usando a função lambda
            resultado = verifica_par_ou_impar(numero)
            print(f"O número {numero} é {resultado}.")
        
        except ValueError:
            # Trata o caso em que a entrada não pode ser convertida para inteiro
            print("Por favor, insira um número inteiro válido.")

if __name__ == "__main__":
    main()
````

### Q7.

Escreva um programa que pergunte ao usuário qual operação ele deseja: maior ou menor.
Em seguida leia uma quantidade indefinida de valores do usuário, até que o usuário digite o valor zero.
Apresente ao final o maior ou menor dos valores digitados de acordo com a escolha do usuário.

**Sua solução deve incluir pelo menos uma função ```lambda```** 

Exemplo de interação:
```
Opções: (1) maior ou (2) menor?
Opção: 1

Digite os valores de entrada. Digite 0 para finalizar a entrada de valores.
4
12
15
1
0

O maior valor é: 15
```

````python
## Escreva aqui o programa principal 
def main():
    # Define funções lambda para encontrar o maior e o menor valor
    encontrar_maior = lambda valores: max(valores) if valores else None
    encontrar_menor = lambda valores: min(valores) if valores else None

    # Solicita ao usuário qual operação deseja realizar
    escolha = input("Você deseja encontrar o maior ou o menor valor? (Digite 'maior' ou 'menor'): ").strip().lower()

    if escolha not in ['maior', 'menor']:
        print("Escolha inválida. O programa será encerrado.")
        return

    valores = []

    while True:
        try:
            # Solicita ao usuário que insira um valor
            valor = int(input("Digite um valor (0 para sair): "))
            
            if valor == 0:
                # Encerra o loop se o valor for 0
                break

            # Adiciona o valor à lista
            valores.append(valor)
        
        except ValueError:
            # Trata o caso em que a entrada não pode ser convertida para inteiro
            print("Por favor, insira um número inteiro válido.")

    # Verifica a escolha e calcula o resultado
    if escolha == 'maior':
        resultado = encontrar_maior(valores)
        if resultado is not None:
            print(f"O maior valor digitado foi: {resultado}")
        else:
            print("Nenhum valor válido foi digitado.")
    elif escolha == 'menor':
        resultado = encontrar_menor(valores)
        if resultado is not None:
            print(f"O menor valor digitado foi: {resultado}")
        else:
            print("Nenhum valor válido foi digitado.")

if __name__ == "__main__":
    main()
````
